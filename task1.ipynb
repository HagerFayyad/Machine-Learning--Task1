```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

data = pd.read_csv("salaries (1).csv")
data.head()
data
```

```python
data.head(10)
```

```python
data.tail(5)
```

```python
input=data.drop('salary_more_then_100k',axis='columns')
input
```

```python
target=data['salary_more_then_100k']
target
```

```python
la_company=LabelEncoder()
la_job	=LabelEncoder()
la_degree=LabelEncoder()
```

```python
input['company_n']=la_company.fit_transform(input['company'])
input['job_n']=la_company.fit_transform(input['job'])
input['degree_n']=la_company.fit_transform(input['degree'])
```

```python
input
```

```python
input_n=input.drop(['company','job',	'degree'],axis='columns')
input_n
```

```python
target
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(input_n,target)
```

```python
import pandas as pd

input_data = [[2, 1, 0]]  # [company, job, degree]
input_df = pd.DataFrame(input_data, columns=['company_n', 'job_n', 'degree_n'])


prediction = model.predict(input_df)
print( prediction)
```
